```python
import pandas as pd
```

```python
cornwall_waterflow_df = pd.read_csv("./Cornwall_data.csv")
cornwall_waterflow_df
```

```python
DateTime_df = cornwall_waterflow_df.iloc[6:,0]
DateTime_df = pd.to_datetime(DateTime_df, format="%Y/%m/%d %H:%M").reset_index(drop=True)
DateTime = DateTime_df.to_list()
DateTime
```

```python
Dates = DateTime_df.dt.strftime("%Y-%m-%d").to_list()
Dates
```

```python
# Times_df = DateTime_df.dt.time
Times = DateTime_df.dt.strftime("%H:%M").to_list()
Times
```

```python
waterlevel_df = cornwall_waterflow_df.iloc[6:,1].astype(float).reset_index(drop=True)
waterlevel = waterlevel_df.to_list()
```

```python
discharge_dict = pd.read_csv("./cornwall_discharge_cleaned.csv").iloc[:,1:].to_dict(orient="list",)
# print(pd.isna(discharge_dict["Jan"][-1]))
```

```python
discharge_dict = pd.read_csv("./cornwall_discharge_cleaned.csv").iloc[:,1:].to_dict(orient="list",)
# print(discharge_dict)

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
months_num = {month: index+1 for index, month in enumerate(months)}

discharge_dates = {}

for k,v in discharge_dict.items():
    for i, val in enumerate(v):
        if pd.notna(val):
            date = f"2024-{months_num[k]:02d}-{i+1:02d}"
            discharge_dates[date] = float(val)

print(discharge_dates)
```

```python
discharge_values = []

for i in range(len(Dates)):
    discharge_values.append(discharge_dates[Dates[i]])

```

```python
station = ["Cornwall"] * len(DateTime)
station_code = [14870] * len(DateTime)
data = {
    "Station Name": station,
    "Station Code": station_code,
    "Date": Dates,
    "Time": Times,
    "DateTime": DateTime,
    "Water Level (meters) - Niveau d'eau (mÃ¨tres)": waterlevel,
    "Discharge (cubic meters per second) - DÃ©bit (mÃ¨tres cubes par seconde)": discharge_values
}
```

# Final dataframe

```python
cornwall_final_df = pd.DataFrame(data)
cornwall_final_df
```

```python
cornwall_final_df.to_csv("./clean/cornwall_final_data.csv", index=False)
```

## Aggregate per day

```python
agg_cornwall = cornwall_final_df.groupby('Date').agg({
    'Water Level (meters) - Niveau d\'eau (mÃ¨tres)': 'mean',
    # 'Discharge (cubic meters per second) - DÃ©bit (mÃ¨tres cubes par seconde)': 'mean'
}).reset_index()

agg_cornwall['Discharge (cubic meters per second) - DÃ©bit (mÃ¨tres cubes par seconde)'] = agg_cornwall['Date'].map(lambda x : discharge_dates[x])

agg_cornwall
```

```python
agg_cornwall.to_csv("./clean/cornwall_agg_data.csv", index=False)
```
